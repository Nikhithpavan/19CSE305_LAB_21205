```python
import numpy as np
import matplotlib.pyplot as plt
def step(x):
    return np.where(x >= 0, 1, 0)

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
W = np.array([10, 0.2, -0.75])
alpha = 0.05
epochs = 1000
convergence_error = 0.002
error_list = []
for epoch in range(epochs):
    error = 0
    for i in range(len(X)):
        
        y_pred = step(np.dot(X[i], W[1:]) + W[0])  
        
        W[1:] += alpha * (y[i] - y_pred) * X[i]
        W[0] += alpha * (y[i] - y_pred)  
        
        error += (y[i] - y_pred) ** 2
    
    error /= len(X)
    error_list.append(error)
    
    if error <= convergence_error:
        print("Converged after", epoch+1, "epochs")
        break


plt.plot(range(1, epoch+2), error_list)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error vs Epochs')
plt.show()
```
